### Evaluation metrics

The final step is to evaluate the performance of the algorithm. This step is particularly important to compare how well different algorithms perform on a particular dataset. For regression algorithms, three evaluation metrics are commonly used:
1. Mean Absolute Error (MAE) is the mean of the absolute value of the errors.
2. Mean Squared Error (MSE) is the mean of the squared errors.
3. Root Mean Squared Error (RMSE) is the square root of the mean of the squared errors.

Refere your self equations for all of those regression algorithm evaluation matrics.

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

You can see that the value of root mean squared error is 4.19, which is more than 10% of the mean value of the percentages of all the temperature i.e. 22.41. This means that our algorithm was not very accurate but can still make reasonably good predictions.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = dataset[['MinTemp', 'MeanTemp']]  # Features
y = dataset['MaxTemp']  # Target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
mlr_model = LinearRegression()
mlr_model.fit(X_train, y_train)
```

```python
y_pred_mlr = mlr_model.predict(X_test)
mse_mlr = mean_squared_error(y_test, y_pred_mlr)
r2_mlr = r2_score(y_test, y_pred_mlr)

print('Multiple Linear Regression:')
print('Mean Squared Error:', mse_mlr)
print('R-squared:', r2_mlr)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred_mlr)
plt.xlabel("Actual MaxTemp")
plt.ylabel("Predicted MaxTemp")
plt.title("Multiple Linear Regression: Actual vs. Predicted")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(degree=2)  # Adjust the degree as needed
X_poly = poly_features.fit_transform(X)
```

```python
X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(X_poly, y, test_size=0.2, random_state=0)
```

```python
plr_model = LinearRegression()
plr_model.fit(X_train_poly, y_train_poly)
```

```python
y_pred_plr = plr_model.predict(X_test_poly)
mse_plr = mean_squared_error(y_test_poly, y_pred_plr)
r2_plr = r2_score(y_test_poly, y_pred_plr)

print('\nPolynomial Linear Regression:')
print('Mean Squared Error:', mse_plr)
print('R-squared:', r2_plr)
```

```python
plt.scatter(y_test_poly, y_pred_plr)
plt.xlabel("Actual MaxTemp")
plt.ylabel("Predicted MaxTemp")
plt.title("Polynomial Linear Regression: Actual vs. Predicted")
plt.show()
```
